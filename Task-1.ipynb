1. Load Dataset, then see the header 

```python
import pandas as pd
data = pd.read_csv('movie_metadata.csv')
data.head(10)
```

2. The duration column contains data on the film length. How many missing values are there

3. Replace the missing values in the duration column with the median value for this column.

```python
temp = data['duration'].dropna()
```

```python
data['duration'] = data['duration'].fillna(median)
```

two decimal places.

```python
avg = np.mean(data['duration'])
print("{avg:.2f}".format(avg = avg))
```

5. Create a movie_duration_category column, which will contain three categories
depending on the film length:
- Category "1. <90" if the film is less than 90 minutes long
- Category "2. 90–120" if the film is between 90 minutes and two hours long (inclusively)
- Category "3. >120" if the film is more than two hours long

```python
data.head()
```

```python
category = []
for a in data['duration']:
    if (a<90) : category.append("<90")
    elif (a>=90 and a<=120): category.append("90-120")
    else : category.append(">120")
    
data['movie_duration_category'] = category
```

```python
data
```

6. Build a summary table for films released after 2000 (inclusively), to list the numbers of
films:
- Table rows: year
- Table columns: movie duration category ("<90", "90–120", ">120")

```python
filtered_data = data[data.title_year>=2000][['title_year','movie_duration_category']]
filtered_data['title_year'] = filtered_data['title_year'].astype(int)
filtered_data['count'] = 1
```

8. The plot_keywords column holds keywords characterizing the film's plot. Using the data
in this column, create a column called movie_plot_category, to contain four categories
depending on the key words in the column:
- Category "love_and_death" if the keywords include both "love" and "death"
- Category "love" if the keywords include the word "love"
- Category "death" if the keywords include the word "death"
- Category "other" if the keywords do not meet the conditions above

```python
plot_keyword_temp = []
for keywords in data['plot_keywords']:
    cat_temp = ''
    keywords = str(keywords)
    if keywords != None and 'love' in keywords:
        cat_temp = 'love'
        if 'death' in keywords:
            cat_temp = cat_temp+'_and_death'
    elif 'death' in keywords:
        cat_temp = 'death'
    else: cat_temp = 'other'
    plot_keyword_temp.append(cat_temp)
```

```python
data['movie_plot_category'] = plot_keyword_temp
data[['plot_keywords', 'movie_plot_category']].head(20)
```

9. The imdb_score column shows a viewer rating for the film. Build a table to reflect the
average rating of films depending on which movie_plot_category category they belong to.

```python
plotcategory = data['movie_plot_category'].unique()
avg_table = {}
for a in plotcategory:
    avg_score = data[data.movie_plot_category == a]['imdb_score']
    avg_score = np.average(avg_score)
    avg_table[a] = avg_score
avg_table = pd.DataFrame.from_dict(avg_table, orient = 'index', columns=['average_score'])
avg_table
```

10. What is the average rating of films in the "love" category? Give the answer as a floatingpoint figure rounded to two decimal places.


```python
print("{:.2f}".format(avg_table['average_score']['love']))
```
