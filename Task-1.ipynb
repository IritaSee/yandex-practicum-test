2. The duration column contains data on the film length. How many missing values are there
in this column?

```python
data['duration'].isna().sum()
```

3. Replace the missing values in the duration column with the median value for this column.

```python
temp = data['duration'].dropna()
```

```python
import numpy as np
median = np.median(temp)
print(median)
```

```python
data['duration'] = data['duration'].fillna(median)
```

```python
data['duration'].isna().sum()
```

4. What is the average film length? Give the answer as a floating-point figure rounded to
two decimal places.

```python
avg = np.mean(data['duration'])
print("{avg:.2f}".format(avg = avg))
```

5. Create a movie_duration_category column, which will contain three categories
depending on the film length:
- Category "1. <90" if the film is less than 90 minutes long
- Category "2. 90–120" if the film is between 90 minutes and two hours long (inclusively)
- Category "3. >120" if the film is more than two hours long

```python
data.head()
```

```python
category = []
for a in data['duration']:
    if (a<90) : category.append("<90")
    elif (a>=90 and a<=120): category.append("90-120")
    else : category.append(">120")
    
data['movie_duration_category'] = category
```
