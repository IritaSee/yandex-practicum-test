2. Highlight user cohorts based on the week of registration in the application. The cohort
identifier should be the week ordinal (for example, the week from July 29 to August 4
should have identifier 31).

```python
ordinal_start = 31
#temporary
s = pd.to_datetime(data.event_date)
temp = pd.DataFrame([])
day = 1
for i, g in data.groupby([(s - s[0]).astype('timedelta64[D]')]):
        g['id'] = ordinal_start
        #print (g)
        day = day+1
        if day == 7:
            ordinal_start = ordinal_start + 1
            day = 1
        temp = pd.concat([temp,g])
data = temp
data
```
